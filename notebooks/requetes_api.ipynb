```python
from sklearn.preprocessing import StandardScaler
df_movies_tfidf = pd.concat(
    [df_movies.reset_index(drop=True),
     df_overview.reset_index(drop=True)],
    axis=1
)

df_final= df_movies_tfidf.drop(['overview', 'release_date', 'vote_average', 'popularity',
       'adult', 'backdrop_path', 'original_language', 'original_title',
       'poster_path', 'video', 'vote_count'], axis= 1)

features_cols= [col for col in df_final.columns if col not in ['movie_id', 'title']]
X= df_final[features_cols]
scaler= StandardScaler()
X_scaled= scaler.fit_transform(X)
df_scaled= pd.DataFrame(X_scaled, columns=features_cols)
```

```python
from sklearn.neighbors import NearestNeighbors

knn= NearestNeighbors(n_neighbors= 11, metric= 'cosine', algorithm= 'brute')

knn.fit(X_scaled)
```

```python
def reco_movie(title, n_recommendations= 10):
    try:
        idx= df_final[df_final['title']== title].index[0]
    except:
        return f'Film {title} non trouv√©'
    
    movie_features= X_scaled[idx].reshape(1,-1)
    
    distances, indices= knn.kneighbors(movie_features, n_neighbors= n_recommendations)
    
    indices= indices[0][1:]
    distances= distances[0][1:]
    
    df_reco= df_final.iloc[indices][['id','title']].copy()
    df_reco['similarity_score']= 1- distances
    
    return df_reco
```
